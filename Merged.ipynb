```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import folium
import seaborn as sns
```

```python
df_rests=pd.read_csv('rests.csv')
df_init=pd.read_csv('initial.csv')
```

```python
df_rests.rename(columns={'Unnamed: 0':'CAMIS'},inplace=True)
df_rests.head()
```

```python
df_init.drop(columns=['Unnamed: 0'],inplace=True)
df_init['DATE']=pd.to_datetime(df_init['DATE'])
df_init.head()
```

```python
df_merged=df_init.merge(df_rests[['CAMIS','DBA','CUISINE DESCRIPTION','is_chain','Community Board','BORO']],on='CAMIS')
df_merged.head()
```

```python
df_merged.shape
```

```python
df_merged.to_csv('merged.csv')
```
